```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "Product": ["Laptop","Chair","Phone","Laptop","Chair","Phone","Tablet","Laptop","Tablet","Chair"],
    "Quantity": [2,5,3,1,2,4,3,2,1,6],
    "UnitPrice": [800,50,500,800,50,500,300,800,300,50],
    "Region": ["North","South","East","West","North","South","East","North","West","South"]
})

data
```

```python
data["Total_sales"]=data["Quantity"]*data["UnitPrice"]
data
```

```python
sales_per_product=data.groupby("Product")["Total_sales"].sum()
sales_per_product
```

```python
sales_per_product.plot(kind="bar")
plt.show()
```

```python
sales_per_region=data.groupby("Region")["Total_sales"].sum()
sales_per_region
```

```python
sales_per_region.plot(kind="pie",autopct="1.1f%%")
plt.ylabel("")
plt.show()
```

```python
print("Total Revenue =", data["Total_sales"].sum())
print("Best Product =", sales_per_product.idxmax())
print("Best Region =", sales_per_region.idxmax())
```
